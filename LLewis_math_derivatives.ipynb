```python
import numpy as np
import matplotlib.pyplot as plt

#sympy - symbolic math in Python
import sympy as sym
import sympy.plotting.plot as symplot

#create symbolic variables in sympy - sympy allows us to compute derivatives
x = sym.symbols('x')

# create a function: 2 multiply by x to the second power
#compute the derivative manually: 2 multiply by 2 (the power), then subtract 1 from the power 2 - 1.
# result:  4x
fx = 2*x**2

#compute its derivative
df = sym.diff(fx, x)

#print the formula and print the result 4x
print(fx)
print(df)

#Now plot the function and the derivative for the simple polynomial
symplot(fx,(x,-4,4), title='The Function')
plt.show()

symplot(df,(x,-4,4), title='The Derivative of the Function')
plt.show()

#*******************************************
#repeat with relu and sigmoid

#create symbolic functions

# The ReLU function
relu = sym.Max(0,x)

#The Sigmoid function
sigmoid = 1 / (1+sym.exp(-x))  #this formula gives me that 'S' looking shape

#graph the functions
p = symplot(relu,(x,-4,4), label='ReLU', show=False, line_color='blue')
p.extend(symplot(sigmoid,(x,-4,4), label='Sigmoid', show=False, line_color='red'))
p.legend = True
p.title ='The Functions'
p.show()

#Now graph the derivatives for the ReLU and Sigmoid functions
p = symplot(sym.diff(relu),(x,-4,4),label='df(ReLU)', show=False, line_color='blue')
p.extend(symplot(sym.diff(sigmoid),(x,-4,4),label='df(Sigmoid)', show=False, line_color='red'))
p.legend = True
p.title = 'The Derivatives'
p.show()

```

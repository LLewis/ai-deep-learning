Computing natural log "ln" using numpy

```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt  #graph visualization

#define a set of points to evaluate
#plot 200 linear spaced numbers that are between .0001 - 1
x = np.linspace(.0001,1,200)

#compute their log
logx = np.log(x);

#plot the logx result using pyplot
fig = plt.figure(figsize=(10,4))

#increase font size . 
plt.rcParams.update({'font.size':15})

plt.plot(x,logx, 'ks-', markerfacecolor='w')
plt.xlabel('x')
plt.ylabel('log(x)')
plt.show()


#Demonstrates how log and exp are inverses

#redefine with fewer points
x = np.linspace(.0001, 1, 10)

#log and exp inverses of each other
logx = np.log(x)
expx = np.exp(x)

#plot 
plt.plot(x,x, color=[.8,.8,.8])
plt.plot(x, np.exp(logx), 'o', markersize = 8)   #cancels out log(expx). therefoe both are equal to  plotting x, x
plt.plot(x, np.log(expx), 'x', markersize = 8)   #cancels out exp(logx). 
plt.xlabel('x')
plt.ylabel('f(g(x)) aka log(x)')
plt.legend(['unity', 'exp(log(x))', 'log(exp(x))'])
plt.show()

```

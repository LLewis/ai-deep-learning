```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras

# Define the model
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mse')

# Define the data
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0])
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0])

# Train the model
model.fit(xs, ys, epochs=300)

# Predict the value
x_prediction = 10.0
y_prediction = model.predict([x_prediction])

# Plot the predicted value and actual value
plt.scatter(xs, ys, label='Actual Values')
plt.scatter(x_prediction, y_prediction, color='red', label='Predicted Value')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Predicted vs Actual Value')
plt.legend()
plt.show()
```

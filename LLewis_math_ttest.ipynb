```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt

#New sub model of the scipy library
import scipy.stats as stats

#parameters
n1 = 30 #samples in dataset 1
n2 = 40 # samples in dataset 2
mu1 = 1 # population mean in dataset 1
mu2 = 2 # poplulation mean in dataset 2

#genearate the data using the random.randn() function
data1 = mu1 + np.random.randn(n1)
data2 = mu2 + np.random.randn(n2)

#then plot two samples
plt.plot(np.zeros(n1), data1, 'ro', markerfacecolor='w', markersize=14)
plt.plot(np.ones(n2), data2, 'gs', markerfacecolor ='w', markersize=14)
plt.xlim([-1, 2])
plt.xticks([0,1], labels=['Group 1', 'Group 2'])
plt.show()
#so is there a statistical significant difference, to find out use ttest in the stats package below to see
```

```python
#t-test via stats package
# _ind = independent samples
#there a different variants of ttest
#data1 - are all the accuracies of model 1
#data2 - are all the accuracies of model 2
t,p = stats.ttest_ind(data1,data2)

#displays as a tuple
#data2 has a larger mean value - it's value is 0.0003186 than
#data1 -3.79314
#if reverse value data2, data1 - data1 has the same value but positive instead of negative

print("data1" ,t)
print("data2" ,p)

```

```python
#common way to show t-test results in a plot
fig = plt.figure(figsize=(10,4))
plt.rcParams.update({'font.size':12}) #change the font size, updating (adding on-fly dictionary font size: 12)

#adds an x axis for ro (red circles) and bs (blue squares) 0+np, 1+np
plt.plot(0+np.random.randn(n1)/15, data1, 'ro',markerfacecolor='w', markersize=14)
plt.plot(1+np.random.randn(n2)/15, data2, 'bs', markerfacecolor='w', markersize=14)
plt.xlim([-1,2])
plt.xticks([0,1],labels=['Group 1', 'Group 2'])

#set the title to include the t-value and p-value
plt.title(f't = {t:.2f}, p={p:.3f}')

plt.show()

#THE P VALUE MUST BE BELOW .05 IN ORDER FOR THE RESULT TO BE SIGNIFICANT
```
